ANDHRA LINKS AND ROUTES

```python

# Initialize the WebDriver
driver_A = webdriver.Chrome()

# Load the webpage
driver_A.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")

time.sleep(3)

# Maximize the browser window
driver_A.maximize_window()

# Initialize WebDriverWait
wait = WebDriverWait(driver_A, 20)

def Andhra_link_route(path):
    LINKS_ANDHRA = []
    ROUTE_ANDHRA = []
    
    for i in range(1, 6):  # Assuming there are up to 5 pages to paginate through
        # Retrieve the route links
        try:
            paths = driver_A.find_elements(By.XPATH, path)
            
            if not paths:
                print(f"No routes found on page {i}.")
                
            for link in paths:
                href = link.get_attribute("href")
                text = link.text
                LINKS_ANDHRA.append(href)
                ROUTE_ANDHRA.append(text)
                print(f"Found link: {href}, text: {text}")
                
        except NoSuchElementException:
            print(f"Could not retrieve elements on page {i}.")
        
        # Pagination
        if i < 5:  # Only attempt to click "Next" if not on the last page
            try:
                # Wait for the pagination element to be present
                pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
                next_page_xpath = f'//div[@class="DC_117_pageTabs " and text()="{i + 1}"]'
                print(f"Attempting to click next page button with XPath: {next_page_xpath}")
                next_button = pagination.find_element(By.XPATH, next_page_xpath)
                
                # Scroll to the element and click
                driver_A.execute_script("arguments[0].scrollIntoView(true);", next_button)
                time.sleep(2)  # Allow time for scrolling
                next_button.click()
                
                # Wait for the new page to load
                time.sleep(3)
                
            except (NoSuchElementException, TimeoutException) as e:
                print(f"Error while navigating to the next page: {e}")
                break
    
    return LINKS_ANDHRA, ROUTE_ANDHRA

# Call the function with the correct XPath
LINKS_ANDHRA, ROUTE_ANDHRA = Andhra_link_route("//a[@class='route']")

# Output the results
print("Links: ", LINKS_ANDHRA)
print("Routes: ", ROUTE_ANDHRA)

# Quit the driver
driver_A.quit()
```
